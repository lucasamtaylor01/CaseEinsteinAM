```python
from pathlib import Path

import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
data_file = Path('..') / 'data' / 'data_processed' / 'SUPERSTORE_TRATADO.csv'
df = pd.read_csv(data_file, sep=',')
```

```python
px.histogram(
    df,
    x='ORDER_DATE',
    y='PROFIT',
    color='CATEGORY',
)
```

```python
df_map = df.groupby('STATE')['PROFIT'].sum().reset_index()

fig = px.choropleth(
    df_map,
    locations='STATE',
    locationmode='USA-states',  
    color='PROFIT',
    scope='usa',  
    color_continuous_scale='RdYlGn'
)

fig.show()
```

```python
df_map = df.groupby('STATE')['PROFIT'].mean().reset_index()

fig = px.choropleth(
    df_map,
    locations='STATE',
    locationmode='USA-states',  
    color='PROFIT',
    scope='usa',  
    color_continuous_scale='RdYlGn'
)

fig.show()
```

```python
fig = px.pie(df, values='PROFIT', names='REGION', title='Lucro por Região')
fig.show()
```

```python
cliente = df['SHIP_MODE'].value_counts().reset_index()
cliente.columns = ['SHIP_MODE', 'COUNT']

fig = px.pie(
    cliente,
    values='COUNT',
    names='SHIP_MODE',
    title='SHIP_MODE mais frequente (por número de pedidos)'
)

fig.show()
```
